```python
def get_table_from_wikipedia(url, pref = 'https://en.wikipedia.org' ):
    
    full_url = f'{pref}{url}' 
    response=requests.get(full_url)
    soup = BeautifulSoup(response.text, 'html.parser')
#     tbls = soup.findAll("table", {"class": "wikitable sortable jquery-tablesorter"})
    tbls = soup.findAll('table')
    
    return soup
```

```python
soup = get_table_from_wikipedia('/wiki/NASDAQ-100')
```

```python
def get_stocks(stock):
    stock_data = yf.Ticker(stock)
    hist = stock_data.history(period="max")
    return hist
```

```python
tikdata = []
for ticker in tables[3].Ticker:
    df = get_stocks(ticker)
    df['Ticker'] = ticker
    tikdata.append(df)
```

```python
nasdaq = pd.concat(tikdata)
```

```python
nasdaq.to_csv('nasdaq.csv')
```

```python
def plot_stock_price(data, col='Close'):
    fig = px.line(data, y = col, x='Date')
    fig.show()
    return data
```
